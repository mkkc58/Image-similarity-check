```python
from PIL import Image
from collections import Counter
import numpy as np
import cv2
import matplotlib.pyplot as plt
from math import sqrt
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

```python
class Similarity():
    
    def imagetoarray(self,img):
        img=np.asarray(img)
        img=img.flatten()
        return img
    
    def img2normalizing(self,img):
        list1=[]
        for i in range(256):
            if i in img.keys():
                list1.append(img[i])
            else:
                list1.append(0)
        return list1
    
    def cosineSim(self,a1,a2):
        sum = 0
        suma1 = 0
        sumb1 = 0
        for i,j in zip(a1, a2):
            suma1 += (i * i)
            sumb1 += (j*j)
            sum += (i*j)
        cosine_sim = sum / ((sqrt(suma1))*(sqrt(sumb1)))
        return cosine_sim
    
    def imagesimilaritycheck(self,img1,img2):
        img1=self.imagetoarray(img1)
        img2=self.imagetoarray(img2)
        
        img1=Counter(img1)
        img2=Counter(img2)
        
        img1=self.img2normalizing(img1)
        img2=self.img2normalizing(img2)
        
        similarity_score=self.cosineSim(img1,img2)
        
        return similarity_score
```

```python
query_image=['Cattle1_1.jpg','Cattle1_2.jpg','Cattle1_3.jpg','Cattle2_1.jpg','Cattle2_2.jpg','Cattle2_3.jpg']
test_image=['test1.jpg','test2.jpg','test3.jpg','test4.jpg']
final_result={}
```

```python
for i in test_image:
    for j in query_image: 
        img1=Image.open(i)
        img2=Image.open(j)
        similarity_check=Similarity()
        score=similarity_check.imagesimilaritycheck(img1,img2)
        final_result[i,j]=score
```

```python
df=pd.Series(final_result).reset_index()
df.columns=['Test Image','Query Image','Similarity Score']
```
